```python
from ultralytics import YOLO
import supervision as sv
import cv2
from google.colab.patches import cv2_imshow

# Replace this with your uploaded image filename
image_path = "images.jpeg"  # üëà Change this!
image = cv2.imread(image_path)
if image is None:
    raise ValueError(f"‚ùå Could not load image: {image_path}")

# Convert BGR to RGB for YOLO
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Load YOLOv8 model
model = YOLO("yolov8n.pt")  # Or your own .pt model

# Run detection
results = model(image_rgb)[0]

# Convert to Supervision format
detections = sv.Detections.from_ultralytics(results)

# Optional tracking (even though it's 1 image‚Äîit aligns with video pipeline)
tracker = sv.ByteTrack()
tracked = tracker.update_with_detections(detections)

# Generate numbered labels like "1. car", "2. truck"
labels = [
    f"{i+1}. {model.model.names[int(cls)]}"
    for i, cls in enumerate(tracked.class_id)
]

# Annotate the image
annotator = sv.BoxAnnotator()
annotated = annotator.annotate(
    scene=image.copy(),
    detections=tracked,
    labels=labels
)

# Show result
cv2_imshow(annotated)
```
